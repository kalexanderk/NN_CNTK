#### Loading data

```python
data_pred = pd.read_csv("Data/prediction.csv", header=0, index_col=None,na_values=['NA'], delimiter=";")
```

```python
data_train = pd.read_csv("Data/training.csv", header=0, index_col=None,na_values=['NA'], delimiter=";")
```

```python
target_train = data_train['target']
```

```python
data_train = data_train.drop(['id', 'target'], axis=1)
```

#### Model

```python
n_cols = data_train.shape[1]
```

```python
#target = target_train[0:10000]
target = target_train
```

```python
target_ = keras.utils.to_categorical(target)
```

```python
#predictors = data_train.iloc[0:10000, ]
predictors = data_train
```

```python
predictors_ = predictors.as_matrix()
```

```python
def get_new_model(n_cols = n_cols):
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(80, activation='relu', input_shape=(n_cols,)))
    model.add(keras.layers.Dense(70, activation='relu'))
    model.add(keras.layers.Dense(60, activation='relu'))
    model.add(keras.layers.Dense(2, activation="softmax"))
    return model
    
```

```python
early_stopping_monitor =  keras.callbacks.EarlyStopping(patience=5)
```

```python
#lr_to_test = [0.1, 1, 10]
lr_to_test = [0.1]
```

```python
model_1 = get_new_model()
```

```python
target = target_train[0:10000]
target_ = keras.utils.to_categorical(target)
predictors = data_train.iloc[0:10000, ]
predictors_ = predictors.as_matrix()
```

```python
for lr in lr_to_test:
    my_optimizer = keras.optimizers.SGD(lr=lr)
    model_1.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ["accuracy"])
    model_1.fit(predictors_, target_, epochs=25, validation_split=0.1, callbacks=[early_stopping_monitor])
```
